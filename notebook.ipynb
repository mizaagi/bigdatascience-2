```python
%pip install pandas seaborn matplotlib numpy -q
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
import getSandP as snp
from datetime import datetime
from concurrent import futures
```

```python
apple_df = pd.read_csv('individual_stocks_5yr/individual_stocks_5yr/AAPL_data.csv')
sorted_df = apple_df.groupby('date')['close'].mean().plot()
plt.title("APPL close prices from 2013 to 2018")
plt.xlabel("Date")
plt.ylabel("Close Price in USD ($)")
plt.xticks(rotation=-20)
sorted_df
```

```python
all_df = pd.read_csv('all_stocks_5yr.csv')
sorted_all_df = all_df.groupby('date')['close'].mean().plot()
plt.title("Average close prices from 2013 to 2018")
plt.xlabel("Date")
plt.ylabel("Close Price in USD ($)")
plt.xticks(rotation=-20)
sorted_all_df
```

As expected, stock prices experience a slow but steady rise from 2013 until 2018, with the notable exception of the price drop in 2016. This drop in the average stock price occurred because of multiple factors, notably the drop in oil prices as well as worry about a declining Chinese economy. As seen in the graph above (of Apple stock price), this drop especially affected them as a technology producer that manufactures many of their parts in China.

```python
plt.figure(figsize=(10, 6))
info = all_df.groupby('high')['low'].sum()
info.plot(kind='line', color='black')
plt.xlabel("High Price")
```

This graph shows that there are many more stocks that have relatively low prices than those that have high prices.

```python
sorted_mean_df = all_df.groupby('Name')['close'].mean()
print(sorted_mean_df)
sorted_mean_df.plot()
```

This graph shows that there is no correlation between alphabetical name and stock closing price, as you would expect.
